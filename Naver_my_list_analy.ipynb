```python
string = urlopen(url)
text = string.read()
#text2 = text.decode('unicode_escape')
text2 = str(text, encoding = 'utf-8')
print(text2)
```

```python
string_soup = BeautifulSoup(text2)
cafe_list = []
rank = []
list_soup = string_soup.find_all("a","cafe_name")
```

```python
i = 1
for txt in list_soup:
    rank.append(str(i))
    cafe_list.append(txt.get_text())
    i= i +1
print(cafe_list)
print(rank)
```
