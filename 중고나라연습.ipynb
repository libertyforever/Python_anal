```

def crawl():
  src = browser.page_source
  soup = BeautifulSoup(src, 'html.parser')
  
  a_list = soup.select('div.inner_list > a')
  for a in a_list:
    if len(a.get_text().strip())>4:
      name_list.append(a.get_text().strip())
  print(name_list)
  return soup.select('div.prev-next > a')

```
